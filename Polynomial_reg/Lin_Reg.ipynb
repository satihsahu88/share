```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X=np.random.rand(100, 1)
```

```python
Y = 4 + 5*X + np.random.randn(100,1)
```

```python
Reg = LinearRegression()
Reg.fit(X,Y)
X_val = np.linspace(0,1,100).reshape(-1,1)
Y_val = Reg.predict(X_val)
```

```python
plt.scatter(X,Y)
plt.scatter(X_val,Y_val, color="r")
plt.show()
```
